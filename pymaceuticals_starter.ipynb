## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
import pylab



# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset
```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
```

```python
# Capomulin, Ramicane, Infubinol, and Ceftamin.

mylist = ['Ramicane','Infubinol', 'Capomulin','Ceftamin']
final_tumor=df.loc[df['Drug Regimen'].isin(mylist)]
final_tumor=final_tumor.groupby('Mouse ID').max()
final_tumor.head(30)
```

```python
ce=final_tumor.loc[final_tumor['Drug Regimen']=='Ceftamin']['Tumor Volume (mm3)']
ca=final_tumor.loc[final_tumor['Drug Regimen']=='Capomulin']['Tumor Volume (mm3)']
r=final_tumor.loc[final_tumor['Drug Regimen']=='Ramicane']['Tumor Volume (mm3)']
i=final_tumor.loc[final_tumor['Drug Regimen']=='Infubinol']['Tumor Volume (mm3)']
i
```

```python
quartiles = ce.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound=lowerq - (1.5*iqr)
upper_bound=upperq + (1.5*iqr)
print(lower_bound, upper_bound)
```

```python
quartiles = ca.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound=lowerq - (1.5*iqr)
upper_bound=upperq + (1.5*iqr)
print(lower_bound, upper_bound)
```

```python
quartiles = r.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound=lowerq - (1.5*iqr)
upper_bound=upperq + (1.5*iqr)
print(lower_bound, upper_bound)
```

```python
quartiles = i.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq
lower_bound=lowerq - (1.5*iqr)
upper_bound=upperq + (1.5*iqr)
print(lower_bound, upper_bound)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

plt.boxplot([ce,ca,r,i])
plt.title('Final Tumor Volume for all Four Treatment Regimens')
plt.ylabel('Drug Regimens')
plt.xlabel('Final Tumor Volume')
plt.show()
```

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
mylist2 = ['s185']
final_tumor2=df.loc[df['Mouse ID'].isin(mylist2)]

final_tumor2
```

```python
mylist3 = ['Capomulin']
weight=df.loc[df['Drug Regimen'].isin(mylist3)]
weight=weight.groupby('Mouse ID').mean()
weight
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
y=weight['Weight (g)']
x=weight['Tumor Volume (mm3)']
```

```python
plt.title("Mouse Weight VS Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Average Tumor Volume")
plt.ylabel("Mouse Weight")

plt.scatter(x, y)
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(x, y)
correlation
```

```python
#linear regression

(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)

regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,1)) + "x + " + str(round(intercept,1))

plt.scatter(x,y)
plt.plot(x,regress_values,"r-")
plt.annotate(line_eq,(6, 0),fontsize=15,color="red")
plt.title("Mouse Weight VS Average Tumor Volume for Capomulin Regimen")
plt.xlabel("Average Tumor Volume")
plt.ylabel("Mouse Weight")
plt.show()
```
